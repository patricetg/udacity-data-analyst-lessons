### Preparatory Step
The cars in this dataset are categorized into one of five different vehicle classes based on size. Starting from the smallest, they are: `{Minicompact Cars, Subcompact Cars, Compact Cars, Midsize Cars, and Large Cars}`.

```python
car_classes = ["Minicompact Cars", "Subcompact Cars", "Compact Cars", "Midsize Cars", "Large Cars"]
v_classes = pd.api.types.CategoricalDtype(ordered=True, categories=car_classes)
fuel_econ["VClass"] = fuel_econ["VClass"].astype(v_classes)
```

```python
color = sb.color_palette()[0]
sb.violinplot(data=fuel_econ,x="VClass",y="displ",color=color)
plt.xticks(rotation=15)
```

```python
sb.boxplot(data=fuel_econ,x="VClass",y="displ",color=sb.color_palette()[0])
plt.xticks(rotation=15)
```

```python
plt.figure( figsize=[16,5] )
color = sb.color_palette()[0]

# left plot
plt.subplot(1,2,1)
ax1= sb.violinplot(data=fuel_econ,x="VClass",y="displ",color=color)
plt.xticks(rotation=15)

# right plot
plt.subplot(1,2,2)
sb.boxplot(data=fuel_econ,x="VClass",y="displ",color=sb.color_palette()[0])
plt.xticks(rotation=15)
plt.ylim(ax1.get_ylim()) # set y-axis limits to be same as left plot, for better comparison
```
